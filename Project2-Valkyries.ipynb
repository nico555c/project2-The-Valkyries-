```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
apple_file = "Data/appleAppData.csv"
apple_df = pd.read_csv(apple_file)
apple_df.head()
```

```python
new_apple = apple_df[['App_Name', 'Primary_Genre', 'Average_User_Rating','Reviews']].copy()
new_apple.rename(columns={'App_Name': 'app_name', 'Primary_Genre': 'primary_genre', 'Average_User_Rating': 'average_user_rating','Reviews':'reviews'}, inplace=True)
new_apple.drop_duplicates(subset='app_name', keep='first', inplace=False)

new_apple.count()
```

```python
new_apple = new_apple.dropna()
new_apple.count()
```

```python
new_apple = new_apple.drop_duplicates()
new_apple.count()
```

```python
new_apple['id'] = new_apple.index
new_apple.head()
new_apple.set_index('id',inplace=True)
new_apple.head()
```

```python
google_file = "Data/googleplaystore.csv"
google_df = pd.read_csv(google_file)
google_df.head()
```

```python
new_google = google_df[['App', 'Category', 'Rating','Reviews']].copy()
new_google.rename(columns={'App': 'app_name', 'Category': 'category','Rating':'rating', 'Reviews': 'reviews'}, inplace=True)
#new_google.drop_duplicates()
#new_google.head()
new_google.count()
new_google.dtypes
```

```python
new_google = new_google.dropna()
new_google.count()
```

```python
new_google = new_google.drop_duplicates()
new_google.count()
```

```python
new_google['id'] = new_google.index
new_google.head()
new_google.set_index('id',inplace=True)
new_google.head()
```

```python
merge_df = pd.merge(new_apple, new_google, on='app_name')
merge_df.rename(columns={'primary_genre': 'Apple Genre','average_user_rating':'Apple Rating', 'reviews_x': 'Apple Reviews','category':'Google Genre', 'rating': 'Google Rating', 'reviews_y':'Google Reviews'}, inplace=True) 
merge_df.head()

#merge_df.count()
```
